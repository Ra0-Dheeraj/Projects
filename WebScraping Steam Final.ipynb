```python
from selenium import webdriver
from selenium.webdriver.support.select import Select
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
```

```python
driver = webdriver.Chrome(executable_path=r"C:\Users\Dell\Downloads\chromedriver.exe")
```

```python
url="https://store.steampowered.com/games/"
driver.get(url)
```

```python
import requests 
import time
import re
```

```python
Game_Name=[]
Review_Number=[]
Developer=[]
Publisher=[]
OS=[]
Processor=[]
Memory=[]
Graphics=[]
Directx=[]
Review=[]
```

```python
## For going on New Releases Page
driver.get('https://store.steampowered.com/games/#p=0&tab=NewReleases') 

## loop to get first 5 Games
for j in range(1,6):
    
    ## Getting name of games using Xpath
    try:
        name= driver.find_element_by_xpath('/html/body/div[1]/div[7]/div[5]/div[1]/div[3]/div[4]/div/div[1]/div/div[2]/div[2]/div/div[1]/a['+str(j)+']/div[3]/div[1]')
        Game_Name.append(name.text)
    except NoSuchElementException:
        Game_Name.append("NA")
        
    ## Clicking on Game page    
    driver.find_element_by_xpath('/html/body/div[1]/div[7]/div[5]/div[1]/div[3]/div[4]/div/div[1]/div/div[2]/div[2]/div/div[1]/a['+str(j)+']').click()
    #time.sleep(0.5)
    
    ## Reviews
    try:
        str1=driver.find_element_by_class_name('user_reviews').text
        x=re.findall(r'\d+', str1)
        if(len(x)==1):
            Review_Number.append(x[0])
        else:
            Review_Number.append("NA")
    except NoSuchElementException:
        Review_Number.append("NA")
                         
    ### Developer
    try:
        str2=driver.find_element_by_xpath('/html/body/div[1]/div[7]/div[5]/div[1]/div[3]/div[1]/div[4]/div[1]/div[2]/div[1]/div/div[3]/div[3]').text
        x=str2.split(':\n')
        Developer.append(x[-1])
    except NoSuchElementException:
        Developer.append("NA")
    
    
    ### Publisher
    try:
        str3=driver.find_element_by_xpath('/html/body/div[1]/div[7]/div[5]/div[1]/div[3]/div[1]/div[4]/div[1]/div[2]/div[1]/div/div[3]/div[4]').text
        x=str3.split(':\n')
        Publisher.append(x[-1])
    except NoSuchElementException:
        Publisher.append('NA')
    
    ##System Requirements
    try:
        str4=driver.find_element_by_class_name('game_area_sys_req_leftCol').text
       
    except NoSuchElementException:
        str4=driver.find_element_by_class_name('game_area_sys_req_full').text

    x=re.split('\n',str4)

    ## OS
    i=re.search('^OS:',x[2])
    if(i!=None):
        z=re.split('OS: ',x[2])
        OS.append(z[-1])
    else:
        OS.append('NA')

    ##Processor
    i=re.search('^Processor:',x[3])
    if(i!=None):
        z=re.split('Processor: ',x[3])
        Processor.append(z[-1])
    else:
        Processor.append('NA')

    ##Memory
    i=re.search('^Memory:',x[4])
    if(i!=None):
        z=re.split('Memory: ',x[4])
        Memory.append(z[-1])
    else:
        Memory.append('NA')

    ##Graphics
    i=re.search('^Graphics:',x[5])
    if(i!=None):
        z=re.split('Graphics: ',x[5])
        Graphics.append(z[-1])
    else:
        Graphics.append('NA')

    ##DirectX
    i=re.search('^DirectX:',x[6])
    if(i!=None):
        z=re.split('DirectX: ',x[6])
        Directx.append(z[-1])
    else:
        Directx.append('NA')
    
    ###comments
    
    ### Scrolling down the page
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    time.sleep(1)
    
    ### finding comments
    str5=driver.find_elements_by_xpath('//*[@id="Reviews_summary"]//div[@class="content"]')
    l1=[]
    for x in str5:
        l1.append(x.text)
    Review.append(l1)
    
    
    ## Getting back to home page
    driver.get('https://store.steampowered.com/games/#p=0&tab=NewReleases')
```

```python
Review_1=Review[0]
Review_2=Review[1]
Review_3=Review[2]
Review_4=Review[3]
Review_5=Review[4]
```

```python
def fillna(list1):
    for x in range(len(list1),11):
        list1.append('NA')
```

```python
fillna(Review_1)
fillna(Review_2)
fillna(Review_3)
fillna(Review_4)
fillna(Review_5)
```

```python
Game_Review=[Review_1,Review_2,Review_3,Review_4,Review_5]
```

```python
import pandas as pd
```

```python
data=pd.DataFrame({'Game_Name':Game_Name,
                  'Review':Review,
                  'Developer':Developer,
                  'Publisher':Publisher,
                  'OS':OS,
                  'Processor':Processor,
                   'Memory':Memory,
                   'Graphics':Graphics,
                   'Directx':Directx,
                  'Game_Review':Game_Review})
```
